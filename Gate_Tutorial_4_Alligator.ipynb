```python
import requests
import json
from typing import Dict, Any

class MantisAPIClient:
    def __init__(self, base_url: str = "http://20.8.170.217:3004", token: str = None):
        """
        Initialize the Mantis API client.
        
        Args:
            base_url (str): Base URL of the API
            token (str): JWT authentication token
        """
        self.base_url = base_url.rstrip('/')
        self.token = token
        self.headers = {
            'Accept': 'application/json, text/plain, */*',
            'Content-Type': 'application/json;charset=UTF-8',
            'Authorization': f'Bearer {token}' if token else None
        }

    def reconcile_table(self, dataset_id: str, table_id: str, table_data: Dict[str, Any]) -> Dict[str, Any]:
        """
        Send a reconciliation request for a specific table.
        
        Args:
            dataset_id (str): ID of the dataset
            table_id (str): ID of the table
            table_data (dict): Table data to be reconciled
            
        Returns:
            dict: API response
            
        Raises:
            requests.exceptions.RequestException: If the API request fails
        """
        endpoint = f"/api/reconciliators/mantis/dataset/{dataset_id}/table/{table_id}"
        url = f"{self.base_url}{endpoint}"
        
        try:
            response = requests.post(
                url=url,
                headers=self.headers,
                json=table_data
            )
            response.raise_for_status()
            return response.json()
            
        except requests.exceptions.RequestException as e:
            print(f"Error making API request: {str(e)}")
            raise

    def get_reconciliation_status(self, dataset_id: str, table_id: str) -> Dict[str, Any]:
        """
        Get the status of a reconciliation request.
        
        Args:
            dataset_id (str): ID of the dataset
            table_id (str): ID of the table
            
        Returns:
            dict: Status response
        """
        endpoint = f"/api/reconciliators/mantis/dataset/{dataset_id}/table/{table_id}/status"
        url = f"{self.base_url}{endpoint}"
        
        try:
            response = requests.get(
                url=url,
                headers=self.headers
            )
            response.raise_for_status()
            return response.json()
            
        except requests.exceptions.RequestException as e:
            print(f"Error checking status: {str(e)}")
            raise
```

```python
# Example usage
token = token_manager.get_token()
dataset_id = dataset_id
table_id = table_id
table_name = table_name

# Initialize client
client = MantisAPIClient(token=token)

try:
    # Send reconciliation request
    result = client.reconcile_table(dataset_id, table_id, table_data)
    print("Reconciliation request result:", result)
    
except requests.exceptions.RequestException as e:
    print(f"An error occurred: {str(e)}")
```
